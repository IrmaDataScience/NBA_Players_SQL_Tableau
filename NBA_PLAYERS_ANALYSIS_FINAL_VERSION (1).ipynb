```python
query = """
WITH top_100_players AS (
    SELECT player_name, player_height, player_weight, pts,
           ROW_NUMBER() OVER (ORDER BY pts DESC) AS rank
    FROM nba_data
)
SELECT 
    'Average BMI' AS stat_type, 
    ROUND(AVG(player_weight / (player_height / 100) / (player_height / 100)), 2) AS bmi,
    CASE 
        WHEN ROUND(AVG(player_weight / (player_height / 100) / (player_height / 100)), 2) < 18.5 THEN 'Underweight'
        WHEN ROUND(AVG(player_weight / (player_height / 100) / (player_height / 100)), 2) BETWEEN 18.5 AND 24.9 THEN 'Normal weight'
        WHEN ROUND(AVG(player_weight / (player_height / 100) / (player_height / 100)), 2) BETWEEN 25 AND 29.9 THEN 'Overweight'
        ELSE 'Obese'
    END AS bmi_category
FROM top_100_players
WHERE rank <= 100
UNION
SELECT 
    'Min BMI' AS stat_type, 
    ROUND(MIN(player_weight / (player_height / 100) / (player_height / 100)), 2) AS bmi,
    CASE 
        WHEN ROUND(MIN(player_weight / (player_height / 100) / (player_height / 100)), 2) < 18.5 THEN 'Underweight'
        WHEN ROUND(MIN(player_weight / (player_height / 100) / (player_height / 100)), 2) BETWEEN 18.5 AND 24.9 THEN 'Normal weight'
        WHEN ROUND(MIN(player_weight / (player_height / 100) / (player_height / 100)), 2) BETWEEN 25 AND 29.9 THEN 'Overweight'
        ELSE 'Obese'
    END AS bmi_category
FROM top_100_players
WHERE rank <= 100
UNION
SELECT 
    'Max BMI' AS stat_type, 
    ROUND(MAX(player_weight / (player_height / 100) / (player_height / 100)), 2) AS bmi,
    CASE 
        WHEN ROUND(MAX(player_weight / (player_height / 100) / (player_height / 100)), 2) < 18.5 THEN 'Underweight'
        WHEN ROUND(MAX(player_weight / (player_height / 100) / (player_height / 100)), 2) BETWEEN 18.5 AND 24.9 THEN 'Normal weight'
        WHEN ROUND(MAX(player_weight / (player_height / 100) / (player_height / 100)), 2) BETWEEN 25 AND 29.9 THEN 'Overweight'
        ELSE 'Obese'
    END AS bmi_category
FROM top_100_players
WHERE rank <= 100
"""

result = ps.sqldf(query, locals())

print(result)
```

```python
query = """
WITH top_100_players AS (
    SELECT player_name, player_height, player_weight, pts,
           ROW_NUMBER() OVER (ORDER BY pts DESC) AS rank
    FROM nba_data
)
, bmi_categories AS (
    SELECT 
        player_name,
        ROUND(player_weight / (player_height / 100) / (player_height / 100), 2) AS bmi,
        CASE 
            WHEN ROUND(player_weight / (player_height / 100) / (player_height / 100), 2) < 18.5 THEN 'Underweight'
            WHEN ROUND(player_weight / (player_height / 100) / (player_height / 100), 2) BETWEEN 18.5 AND 24.9 THEN 'Normal weight'
            WHEN ROUND(player_weight / (player_height / 100) / (player_height / 100), 2) BETWEEN 25 AND 29.9 THEN 'Overweight'
            ELSE 'Obese'
        END AS bmi_category
    FROM top_100_players
    WHERE rank <= 100
)
SELECT 
    bmi_category,
    COUNT(*) AS count,
    ROUND((COUNT(*) * 100.0 / (SELECT COUNT(*) FROM top_100_players WHERE rank <= 100)), 2) AS percentage
FROM bmi_categories
GROUP BY bmi_category
"""
result = ps.sqldf(query, locals())

print(result)
```
